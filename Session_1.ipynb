## Properties of Asymptotic Notation


### General Properties
If $f(n)$ is $O(g(n))$ then $a*f(n)$ is $O(g(n))$

### Reflexive
If $f(n)$ is given then $f(n)$ is $O(f(n))$

### Transitive
If $f(n)$ is $O(g(n))$ and $g(n)$ is $O(h(n))$ then $f(n) = O(h(n))$

### Symmetric
If $f(n)$ is $O(g(n))$ then $g(n)$ is $O(f(n))$

###     Transpose Symmetric
If $f(n)$ is $O(g(n))$ then $g(n)$ is $\Omega(f(n))$

### Other things
If $f(n) = O(g(n))$ and $f(n) = \Omega(g(n))$, $f(n) = \Theta(g(n))$

If $f(n) = O(g(n))$ and $d(n) = O(e(n))$ then $f(n) + d(n) = O(max(g(n),e(n)))$

If $f(n) = O(g(n))$ and $d(n) = O(e(n))$ then $f(n)*d(n) = O(g(n) * e(n))$
